```python
import pandas as pd
df = pd.read_csv('../dataset/train_lyrics_1000.csv')
lyrics = ' '.join(df['lyrics'])
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, STOPWORDS

from scipy.misc import imread

mask = imread('./music-note-on-white.png', flatten=True)


wordcloud = WordCloud( 
                      stopwords=STOPWORDS.add('nigga'),
                      background_color='white',
                      width=5400,
                      height=4800,
                      mask=mask
                     )
wordcloud.generate(lyrics)

plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./lyrics_musicnote_cloud.png', dpi=300)
plt.show()
```

```python
is_sad =  df['mood']=='sad'
df_sad = df[is_sad]
is_happy =  df['mood']=='happy'
df_happy = df[is_happy]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, STOPWORDS

from scipy.misc import imread

mask = imread('./cloud.png', flatten=True)

text_happy = df_happy.lyrics[0]
wordcloud = WordCloud( 
                      stopwords=STOPWORDS.add('nigga'),
                      background_color='White',
                      width=8400,
                      height=4800,
                      mask=mask
                     )
wordcloud.generate(text_sad)

plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./lyrics_sad_cloud.png', dpi=300)
plt.show()
```

```python
text_happy = df_happy.lyrics[1]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text_happy)

# Display the generated image:
plt.imshow(wordcloud, interpolation='lanczos')
plt.axis("off")
plt.show()
```
